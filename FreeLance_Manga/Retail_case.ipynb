```python
Я рассмотрел поставленную вами задачи и дополнительную информацию о ней в сети интернет.
Мои вывод такие - необходимо сравнить дельту затрат в период 18 недель по торговой сети.
Акция длиться 6 недель и её модель основанна на стимуляции повышеения затратности в сети.
Следовательно, колличествая оценка возможна только в случае сравнения за аналогичный период.

Нам потребуеться срез данных за 18 недель: 6 недель по акции, 6 недель акции и 6 недель после акции.
Оценка эффективности опирается на дельту затратности по акции минус дельта из медианы до и после неё.
Так например, если совокупная затратность за 6 недель до акции равна Х1, по акции Х2, а после Х3 тогда:

Абс.формула оценки будет иметь вид: Х2 - (Х1 - Х3)\2 и эта оценка будет количественным значением прибыли.
Полученное значение будет абсолютным и не будет учитывать других факторов влияния по сети - потребуеться:

1. оценить акцию в срезе 18 недель по фокус группам.
2. оценить акцию в срезе 18 недель по колличественным параметрам.

Считаю неоюходимым рассматривать следующие фокус группы:
    
1. группу с чеком 555+ рублей.
2. группу с чеком 555+ рублей получивших ластик 16 раз.
3. группу с чеком 555+ рублей получивших ластик около 50 раз.
4. группу с чкеом 555+ рублей получивших ластик 100 раз и более.

Данное утверждение основанно на том, что в среднем требуеться 100 подходов до полного набора ластиков.

Считаю необходимым дополнить выборку следующими параметрами.

Основная фокус группа №1, далее group_1

1. численность группы с чеком 555+ рублей в период за 6 недель до акции.
2. численность группы с чеком 555+ рублей в период после 6-ти недель акции.
2. численность группы с чеком 555+ рублей в период 6-ти недель во время акции.

Основная фокус группа №2, далее group_2

1. численность группы с чеком 400+ рублей в период за 6 недель до акции.
2. численность группы с чеком 400+ рублей в период после 6-ти недель акции.
2. численность группы с чеком 400+ рублей в период 6-ти недель во время акции.

Основная фокус группа №3, далее group_3

1. численность группы с чеком 100+ рублей в период за 6 недель до акции.
2. численность группы с чеком 100+ рублей в период после 6-ти недель акции.
2. численность группы с чеком 100+ рублей в период 6-ти недель во время акции.

Главная группа, далее main_group

1. численность группы с чеком 555+ рублей в период 6-ти недель во время акции, получивших 1 раз.

Целевая фокус группа №1, далее focus_group_1

1. численность группы с чеком 555+ рублей в период 6-ти недель во время акции, получивших ластик 16 раз.

Целевая фокус группа №2, далее focus_group_2

1. численность группы с чеком 555+ рублей в период 6-ти недель во время акции, получивших ластик около 50 раз.

Целевая фокус группа №3, далее focus_group_3

1. численность группы с чеком 555+ рублей в период 6-ти недель во время акции, получивших ластик 100 раз.

Группа отказа №1, далее cancel_group_1

1. численность группы с чеком 555+ рублей в период за 6 недель до акции, не бравшая ластики.
2. численность группы с чеком 555+ рублей в период после 6-ти недель акции, не бравшая ластики.
2. численность группы с чеком 555+ рублей в период 6-ти недель во время акции, не бравшая ластики.

Группа отказа №2, далее cancel_group_2

1. численность группы с чеком 555+ рублей в период за 6 недель до акции, по магазинам отказа.
2. численность группы с чеком 555+ рублей в период после 6-ти недель акции, по магазинам отказа.
2. численность группы с чеком 555+ рублей в период 6-ти недель во время акции, по магазинам отказа.

# многочисленный отзывы показывают на отказ кассиров выдавать ластики, один заргестрированный отказ == 1000.

Параметры затратности, далее period_profit_1,2,3 

1. Затратность по всем указанным группам.

Дополнительные параметры, далее period_sales_1,2,3

1. Общее кол-во всех ластиков полученных каждой из групп (если это было возможно) за период акции.

Дополнительные коофициенты, далее period_indicator_1,2,3 и sales_indicator_1,2,3

1. Соотношение: число ластиков\ кол.во человек в группе.
2. Соотношение: число акционных товаров \ кол.во человев в группе.
    
Частоты генерации чеков в группе.

На самом деле спорный и неодназначный индикатор, который я решил оспаривать.Это магазин FMCG а не
- лавка игрушек. Так что частоты генерации чеков не могут быть прямо объяснены акцией с ластиками.

Надо смотреть прирост группы с чеком ниже 555 рублей в сторону акции и мы этот индикатор считаем,
считаем по честной формуле Y2 - (Y1 - Y3\2) где: 
    Y1 - число клиентов с чеком ниже 555 рублей до акции
    Y3 - число клиентов с чеком ниже 555 рублей после акции
    Y2 - число клиентов с чеком ниже 555 рублей в период акции.
    Получим вполне себе известное число, отпилим от него минус и скажем директору - мы приросли.
    В свою очередь прирост прибыли по группе с чеком от 555 рублей с оглядкой на новых клиентов
    посчитаем по тому же принципу: Z2 - (Z1-Z2\2) где Z1,2,3 затраты групп за аноогичный период.
        
Эту модель коНечно можно и нужно разворачивать вправо, обогощать данные и считать, но:
во первых меня просили просто подумать, а во-вторых этот анализ стоил мне времени.
Время, деньги. У меня 2:07 ночи на часах, иначе Я бы не успел другое до выходных.
А по выходным лично я смотрю аниме, читаю учебники и валяюсь на дивание - сплю.

Так что остановимся на текущем варианте. 
Ниже будет рабочий облик витрины данных.
```
