```python
from collections import Counter
import re
```

```python
import pandas as pd
```

```python
def glossary(path,common_score):
    """
    Фукнция читает книгу.
    Фукнция возвращает список самых повторяем слов в книге.
    Фукнция возвращает датафрейм состоящий из признаков Words - слова и Score - число повторений.
    Функция прочитала книжку Боло Восстание на русском языке = Хеллбор оказался даже не в первой сотне.
    """
    words = re.findall(r'\w+', open(path, encoding = 'utf-8').read().lower())
    words_list = Counter(words).most_common(common_score)
    df = pd.DataFrame(words_list, columns=['Words','Score'])
    return words_list, df
```
